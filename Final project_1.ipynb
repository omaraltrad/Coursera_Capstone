```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['venue.categories']
    except:
        categories_list = row['categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# filter the category for each row in all extracted dataframe features.

Feature_1['venue.categories'] = Feature_1.apply(get_category_type, axis=1)
Feature_2['venue.categories'] = Feature_2.apply(get_category_type, axis=1)
Feature_3['venue.categories'] = Feature_3.apply(get_category_type, axis=1)
#Feature_4['categories'] = Feature_4.apply(get_category_type, axis=1)
Feature_5['venue.categories'] = Feature_5.apply(get_category_type, axis=1)
Feature_6['venue.categories'] = Feature_6.apply(get_category_type, axis=1)
# Just an example let us examine Feature_5
Feature_1
```

```python
# List the most common Crime categories 
DF_count = DF_Crime[['TYPE', 'YEAR']].groupby(
    ['TYPE'], as_index=False).count().sort_values(
    'YEAR', ascending=False)
DF_count.rename(columns={"YEAR": "Count"},inplace='True')
DF_count
```

```python
# add to the map the expected crime areas
from folium import plugins
from folium.plugins import HeatMap

max_amount = float(DF_count['Count'].max())


hm_wide = HeatMap( list(zip(DF_Crime.Latitude.values, DF_Crime.Longitude.values, DF_count.Count.values)),
                   min_opacity=0.2,
                   max_val=max_amount,
                   radius=17, blur=15, 
                   max_zoom=1).add_to(map_vancouver)

map_vancouver
```
